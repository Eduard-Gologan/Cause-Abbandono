```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
import statsmodels as sm
import sklearn.linear_model as LinearRegression
import scipy.stats as stats

from sklearn.model_selection import train_test_split

 
df=pd.read_csv(r"C:\Users\User\Dropbox\PC\Desktop\WA_Fn-UseC_-HR-Employee-Attrition.csv", sep =';')
df.head()
```

```python
#It's better to know the dataframe info
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.info()
df.describe()
```

```python



plt.figure(figsize=(6,4))
sns.countplot(data=df, x="OverTime", hue="Attrition")
plt.title("Attrition by Overtime")

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="WorkLifeBalance", hue="Attrition")
plt.title("Attrition by Work-life Balance")

plt.savefig("attrition_by_work_life_balance.png", dpi=300)
```

```python
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No":0})


numeric= df[[
    "EnvironmentSatisfaction", 
    "JobSatisfaction", 
    "RelationshipSatisfaction",
    "Attrition"
]]

corr=numeric.corr()


plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="Blues", vmin=-1, vmax=1,square =True)
plt.title("Correlation Matrix of Numeric Features")



df["Attrition"] = df["Attrition"].map({1: "Yes",  0: "No"})
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Attrition", y= "JobSatisfaction")
plt.title("Job Satisfaction vs attrition")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="TrainingTimesLastYear", hue="JobInvolvement", palette="Set2")
plt.title("Training vs Job Involvement")

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="TrainingTimesLastYear", hue="Attrition")
plt.title("Attrition vs Training Times Last Year")
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="JobRole", hue="Attrition", palette="pastel")
plt.title("Attrition Rate by Job Role")
plt.ylabel("Attrition")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("attrition_by_job_role.png", dpi=300)
plt.show()
```
